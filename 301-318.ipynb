```python
def prob_311(n_list):
    return [n + max(n_list) for n in n_list if n > 0]
list1 = [5, -4, 2, -6, 0, 5, -14, 9, 8, 10, 11]
prob_311(list1)
```

```python
def prob_312(input_list):
    new_list = []
    for i in range(0, len(input_list)-1, 2):
        new_list.append(max(abs(input_list[i]), abs(input_list[i+1])))  
    return new_list
input_list = [1, -5, -3, 8, -10, 12, 6, -9]
prob_312(input_list) 
```

```python
def prob_313(input_list):
    for i in range(2, len(input_list), 2):
        input_list[i], input_list[i - 1] = input_list[i - 1], input_list[i]
        return input_list
input_list = [1, 5, 3, 8, 10, 12, 6, 9]
prob_313(input_list)
```

```python
def prob_314(input_list):
    new_list = []
    for num in input_list:
        new_list.append(num) 
        if num > 0:
            new_list.append(0)  
    return new_list
input_list = [1, -5, 3, -8, -10, 12, -6, 9]
prob_314(input_list)
```

```python
def prob_315(input_list, b):
    if b > max(input_list):
        return [num for num in input_list if num > 0]
    else:
        return [num for num in input_list if num < 0]
input_list = [-5, 3, -8, -10, 12, -6, 9]
prob_315(input_list, 5)
```

```python
def prob_316(input_list):
    return [num for num in input_list if abs(num) < max(input_list) + min(input_list) / 2]
input_list = [-5, 2, -8, -10, 12, -6, 9]
output_list = prob_316(input_list)
print(output_list)  
```

```python
def prob_317(input_list):
    max_even_number = max([num for num in input_list if num % 2 == 0])  
    new_list = [input_list[i] + max_even_number for i in range(1, len(input_list), 2)]
    return new_list
input_list = [-5, 3, -8, -10, 12, -6, 9]
prob_317(input_list)
```

```python
def prob_318(input_list):
    negative_nums = [num for num in input_list if num < 0]
    positive_nums = [num for num in input_list if num > 0]
    zeros = [0] * input_list.count(0)
    return negative_nums + zeros + positive_nums
input_list = [-5, 3, -8, 0, -10, 0, 12, -6, 9]
prob_318(input_list)
```
